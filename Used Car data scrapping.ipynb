```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
url ="https://www.olx.in/en/used%20cars_c84"
driver.get(url)
time.sleep(6)
```

```python
price1=driver.find_elements_by_xpath("//span[@class='_89yzn']")
price1
```

```python
price=[]
for p in price1:
    price.append(p.text)
price
```

```python
detail=driver.find_elements_by_xpath("//span[@class='_2TVI3']")
detail
```

```python
year=[]
driven=[]
for i in detail:
    year.append(i.text.split('-')[0])
year
```

```python
model_name=[]
model=driver.find_elements_by_xpath("//span[@class='_2tW1I']")
for m in model:
    model_name.append(m.text)
model_name
```

```python
variant=[]
for i in model_name:
    variant.append(i.split(",")[-1])
variant
```

```python
location=[]
location1=driver.find_elements_by_xpath("//span[@class='tjgMj']")
for l in location1:
    location.append(l.text)
location
```

```python
detail=driver.find_elements_by_xpath("//li[@class='EIR5N']")
detail
```

```python
detail1=[]
for d in detail:
    detail1.append(d.text)
detail1
```

```python
driven=[]
for i in detail:
    driven.append(i.text.split("\n")[-4])
driven
```

```python
driven1=[]
for i in driven:
    driven1.append(i.split('-')[1])
driven1
```

```python
import pandas as pd
car_detail=pd.DataFrame({})
car_detail['Model']=model_name
car_detail['Location']=location
car_detail['Year']=year
car_detail['Driven']=driven1
car_detail['Variant']=variant
car_detail['Price']=price
car_detail
```

```python
import joblib
joblib.dump(car_detail,"car_detail.csv")
```
